```python

dic={
    "A":["B","C"],
    "B":["A","D","E","F"],
    "C":["A","D","F"],
    "D":["B","E","F","C"],
    "E":["B","D","F"],
    "F":["B","D","E","C"]
}
```

```python
def dfs(s,g):
    vis=set()
    st=[]
    st.append(s)
    while(len(st)!=0):
        node=st.pop()
        if(node==g):
            print(node)
            return True
        if(node not in vis):
            vis.add(node)
            print(node)        
            for i in dic[node]:
                st.append(i)
```

```python
dfs("A","F")
```

```python
from collections import deque
def dfs(s,g):
    vis=set()
    st=deque([s]) 
    while(len(st)!=0):
        node=st.popleft()
        if(node==g):
            print(node)
            return True
        if(node not in vis):
            vis.add(node)
            print(node)        
            for i in dic[node]:
                st.append(i)
```

```python
dfs("A","F")
```
