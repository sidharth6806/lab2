```python

dic={
    "A":["B","C"],
    "B":["A","D","E","F"],
    "C":["A","D","F"],
    "D":["B","E","F","C"],
    "E":["B","D","F"],
    "F":["B","D","E","C"]
}
```

```python
def dfs(s,g):
    vis=set()
    st=[]
    st.append(s)
    while(len(st)!=0):
        node=st.pop()
        if(node==g):
            print(node)
            return True
        if(node not in vis):
            vis.add(node)
            print(node)        
            for i in dic[node]:
                st.append(i)
```

```python
dfs("A","F")
```

```python
from collections import deque
def dfs(s,g):
    vis=set()
    st=deque([s]) 
    while(len(st)!=0):
        node=st.popleft()
        if(node==g):
            print(node)
            return True
        if(node not in vis):
            vis.add(node)
            print(node)        
            for i in dic[node]:
                st.append(i)
```

```python
dfs("A","F")
```

```python
# Design a program for dfs:

def dfs_path(start,goal,graph,visited=None):
    if visited is None:
        visited = set()
    print(start)
    if start == goal:
        return True
    visited.add(start)
    for child in graph[start]:
        if child not in visited:
            if dfs_path(child,goal,graph,visited):
                return True
    return False

graph = {
    'A': ['B','C'],
    'B': ['E','D','F'],
    'C': ['D','F'],
    'D': ['E','F'],
    'E':['F']
}
dfs_path('A','F',graph)
```

```python
# Design a code for bfs:

from collections import deque

def bfs_path(graph,start,goal):
    queue = deque([[start]])
    print("queue",queue)
    visited = set([start])

    while queue:
        path = queue.popleft()
        print("path",path)
        node = path[-1]
        print("node",node)

        if node == goal:
            return path

        for neighbour in graph[node]:
            if neighbour not in visited:
                visited.add(neighbour)
                new_path = path + [neighbour]
                print("",new_path)
                queue.append(new_path)
    return None
graph = {
    'A': ['B','C'],
    'B': ['E','D'],
    'C': ['D','F'],
    'D': ['E','F'],
    'E':['F']
}
bfs_path(graph,'A','F')
```
